10 . Compare the SURF and HOG feature descriptors by extracting features from the same image. Visualize and discuss the differences in keypoints and feature representation.

```python
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog

# Load the image
image = cv2.imread('/content/deadpool-wolverine-wallpaper-preview.jpg', cv2.IMREAD_GRAYSCALE)

# ORB Feature Extraction (Alternative to SURF)
orb = cv2.ORB_create()  # Create ORB detector
keypoints_orb, descriptors_orb = orb.detectAndCompute(image, None)
image_orb = cv2.drawKeypoints(image, keypoints_orb, None)

# HOG Feature Extraction
features_hog, hog_image = hog(image, visualize=True, pixels_per_cell=(8, 8), cells_per_block=(2, 2))

# Plot the results
plt.figure(figsize=(10, 5))

# Original Image with ORB Keypoints
plt.subplot(1, 2, 1)
plt.imshow(image_orb, cmap='gray')
plt.title("ORB Keypoints")

# HOG Visualization
plt.subplot(1, 2, 2)
plt.imshow(hog_image, cmap='gray')
plt.title("HOG Features" )

plt.tight_layout()
plt.show()
```
