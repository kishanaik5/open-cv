```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('puppy_dog.jpg')
plt.imshow(image)
```

```python
height, width = image.shape[:2]
translation_matrix = np.float32([[1, 0, 100], [0, 1, 50]]) # Move 100 pixels right and 50 down
translated_image = cv2.warpAffine(image, translation_matrix, (width, height))
plt.imshow(translated_image)
```

```python
scaled_image = cv2.resize(image, None, fx=0.5, fy=0.5,interpolation=cv2.INTER_LINEAR) # Resize to half 
plt.imshow(scaled_image)
```

```python
center = (width // 2, height // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1) # Rotate 45 rotation degrees
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
plt.imshow(rotated_image)
```

```python
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2= np.float32([[10, 100], [200, 50], [100, 250]])
affine_matrix = cv2.getAffineTransform (pts1, pts2)
affine_image=cv2.warpAffine(image,affine_matrix,(width,height))
plt.imshow(affine_image)
```

```python
pts1 = np.float32([[0, 0], [width - 2, 0], [0, height - 3], [width - 1, height - 1]])
pts2 = np.float32([[0, 0], [width - 1, 0], [int(0.58 * width), height - 1],[ int(0.8
* width), height - 1]])
perspective_matrix = cv2.getPerspectiveTransform(pts1, pts2)
perspective_image = cv2.warpPerspective(image, perspective_matrix, (width,height))
plt.imshow(perspective_image)
```
