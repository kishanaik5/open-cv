5 . Show original image and transformed image from different angle

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Read the image
image = cv2.imread('/content/vECHyt.jpg')

# Define source and destination points
src = np.float32([[100, 150], [200, 150], [200, 250], [100, 250]])
dst = np.float32([[50, 100], [250, 100], [250, 300], [50, 300]])

# Compute the homography matrix
homography_matrix, _ = cv2.findHomography(src, dst)

# Warp the image
warped_image = cv2.warpPerspective(image, homography_matrix, (image.shape[1], image.shape[0]))

# Display the images
cv2_imshow(image)
cv2_imshow(warped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
