11 . Implement a region counting algorithm to count the number of distinct regions in a segmented image. Display the original image with the regions labeled by their count.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image and convert to grayscale
img = cv2.imread("/content/deadpool-wolverine-wallpaper-preview.jpg", cv2.IMREAD_GRAYSCALE)

# Apply adaptive thresholding
thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 255, 19)

# Morphological operations to refine objects
kernel = np.ones((5,5), np.uint8)
processed = cv2.medianBlur(cv2.erode(cv2.dilate(thresh, kernel), kernel), 7)

# Count connected components
num_labels, labels = cv2.connectedComponents(processed)

# Display result
plt.imshow(labels, cmap="gray")
plt.title(f"Objects Counted: {num_labels - 1}")
plt.axis("off")
plt.show()

print(f"Objects number is: {num_labels - 1}")
```
